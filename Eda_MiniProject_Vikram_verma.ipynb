```python
#1)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading data
df=pd.read_csv("fifa.csv")
df1=pd.read_csv("fifa_ variable_information.csv")
```

```python
# prepocessing
df.head()
```

```python
#2)
# dropping columns
df1
df.drop(["Photo","Nationality","Flag","Club Logo","Jersey Number"],axis=1,inplace=True)
```

```python
df
```

```python
#3)
# changing data types 
# conversion for Value column
# removng € symbol
df.Value=df.Value.str[1:]
df.Value
```

```python
x=np.where(df.Value.str[-1:]=="M",1000000,1000)
x=pd.Series(x)
x
```

```python
#converting to float
df.Value=df.Value.str.rstrip("M")
df.Value=df.Value.str.rstrip("K")
df.Value=df.Value.astype("Float64")
df.Value
```

```python
#final value column
df.Value=df.Value*x
```

```python
df.Value
```

```python
#converting wage to float
#removing €
df.Wage=df.Wage.str[1:]
df.Wage
```

```python
x=np.where(df.Wage.str[-1:]=="M",1000000,1000)
x=pd.Series(x)
x
```

```python
#converting to float

df.Wage=df["Wage"].str.rstrip("K")
df.Wage=df.Wage.astype("Float64")
df.Wage
```

```python
#final wage column
df.Wage=df.Wage*x
df.Wage
```

```python
# converting Joined to int
df.Joined=df.Joined.str[-4:]
df.Joined.fillna(0,inplace=True)
```

```python
# final joined conversion
df.Joined=df.Joined.astype(int)
df.Joined
```

```python
# df Contract Valid Until 
# dealing with na
df["Contract Valid Until"].fillna(0,inplace=True)
df["Contract Valid Until"]=df["Contract Valid Until"].astype('datetime64')
df["Contract Valid Until"]
```

```python
x=df.Height.str[0]
y=df.Height.str[2:]
```

```python
x=x.astype("float64")
x
```

```python
y=y.astype("float64")
y
```

```python
#final conversion (height in cms)
df.Height=(x*12+y)*2.54
df.Height
```

```python
#Modifying weights
df.Weight=df.Weight.str.rstrip("lbs")
```

```python
#final Conversion to float
df.Weight=df.Weight.astype(float)
df.Weight
```

```python
df.Weight.isnull().sum()
```

```python
# changing release clause variable
df["Release Clause"]
```

```python
#removing €
df["Release Clause"]=df["Release Clause"].str[1:]
df["Release Clause"]
```

```python
x=np.where(df["Release Clause"].str[-1:]=="M",1000000,1000)
x=pd.Series(x)
x
```

```python
# removing M
df["Release Clause"]=df["Release Clause"].str.rstrip("M")
df["Release Clause"]
```

```python
# removing k

df["Release Clause"]=df["Release Clause"].str.rstrip("K")
df["Release Clause"]
```

```python
#Changing datatype
df["Release Clause"]=df["Release Clause"].astype(float)
df["Release Clause"]
```

```python
df["Release Clause"]=df["Release Clause"]*x
df["Release Clause"]
```

```python
#4) dealing with missing data
df.isnull().sum()/len(df)*100
```

```python
# missing data and average imputation
# even though release clause is less than 10% if we remove records with missing release clause data 
# we might lose out on key players
df["Release Clause"].skew()
df["Release Clause"].median()# as the distribution is heaving right skewed
df["Release Clause"]=df["Release Clause"].fillna(df["Release Clause"].median())
```

```python
df.Height.skew()
#not a lot of skew so we can use mean
```

```python
df.Height=df.Height.fillna(df.Height.mean())
```

```python
df.Weight.skew()
#not a lot of skew so we can use mean
```

```python
df.Weight=df.Weight.fillna(df.Weight.mean())
```

```python
df["International Reputation"].skew()
# right skewed so I have picked median
```

```python
df["International Reputation"]=df["International Reputation"].fillna(df["International Reputation"].median())
```
